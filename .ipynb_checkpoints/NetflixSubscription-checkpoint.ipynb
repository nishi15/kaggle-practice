```python
import pandas as pd
import os
data_path = r"E:\Neeshu\kaggle-practice\datasets"
file_name = "Netflix subscription fee Dec-2021.csv"
```

```python
df = pd.read_csv(os.path.join(data_path,file_name))
df["total_cost"] = df["Cost Per Month - Basic ($)"]+df["Cost Per Month - Standard ($)"]+df["Cost Per Month - Premium ($)"]
df["average_total_cost"] = df["total_cost"].mean()
df["deviated_cost"] = df["total_cost"] - df["average_total_cost"]
df.head()
```

**Plotting a line graph for Country name again their Average total cost**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,7))
plt.plot(df["Country"],df["total_cost"],label="total cost",marker="o")
plt.plot(df["Country"],df["average_total_cost"],label="Mean cost",linestyle='dashed',color="green")

plt.xlabel("country name")
plt.xticks(rotation=90) # for rotating x-ticks value to vertically
plt.ylabel("total cost")

# getting maximum cost
max_cost = df.loc[df["total_cost"] == (max(df["total_cost"]))]   
max_cost = list(zip(max_cost["Country"],max_cost["total_cost"]))

# getting minimum cost
min_cost = df.loc[df["total_cost"] == (min(df["total_cost"]))]   
min_cost = list(zip(min_cost["Country"],min_cost["total_cost"]))

# adding text for maximum and minimum
for i, txt in enumerate(max_cost):
    plt.annotate("Maximum "+txt[0], (txt[0],txt[1]))
    
for i, txt in enumerate(min_cost):
    plt.annotate("Minimum "+txt[0], (txt[0],txt[1]))
    
plt.legend()
plt.show()
```
